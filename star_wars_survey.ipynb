# Star Wars Survey

While waiting for Star Wars: The Force Awakens to come out, the team at [FiveThirtyEight](https://fivethirtyeight.com/) became interested in answering some questions about Star Wars fans. For example, does the rest of America realize that Star Wars: The Empire Strikes Back is clearly the best of the bunch? 

To collect the data addressing their questions, they surveyed Star Wars fans using the online tool SurveyMonkey. They received 835 total responses, which are available [here](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey). You can read about it more in [this article](https://fivethirtyeight.com/features/americas-favorite-star-wars-movies-and-least-favorite-characters/).

In this project we will clean and explore this data set. The encoding is ISO-8859-1.

### Overview

To read in and preview the data:

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.shape
```

There is a missing `RespondentID` value, so to keep only not-null values of `RespondentID`:

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
star_wars.head(10)
```

```python
star_wars.shape
```

```python
star_wars.columns
```

### Cleaning and mapping yes/no columns

We can take a look at the columns:
- `Have you seen any of the 6 films in the Star Wars franchise?`
- `Do you consider yourself to be a fan of the Star Wars film franchise?`

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

Both columns contain answers "Yes" or "No" as strings, but there is also NaN where a respondent chose not to answer the question. We can convert each column to Boolean, as Booleans are easier to work with:

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

### Cleaning and mapping checkbox columns

The next six columns represent a single checkbox question:
- `Which of the following Star Wars films have you seen? Please select all that apply.` - whether or not the respondent saw Star Wars: Episode I The Phantom Menace
- `Unnamed: 4` - whether or not the respondent saw Star Wars: Episode II Attack of the Clones
- `Unnamed: 5` - whether or not the respondent saw Star Wars: Episode III Revenge of the Sith
- `Unnamed: 6` - whether or not the respondent saw Star Wars: Episode IV A New Hope
- `Unnamed: 7` - whether or not the respondent saw Star Wars: Episode V The Empire Strikes Back
- `Unnamed: 8` - whether or not the respondent saw Star Wars: Episode VI Return of the Jedi

If a value in a cell is name of the movie, that means the respondent saw the movie. If the value is NaN, the respondent didn't see the movie or didn't answer. We will assume the respondent didn't see the movie.

To convert the columns to Boolean values:

```python
import numpy as np

episodes = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones":True,
    "Star Wars: Episode III  Revenge of the Sith":True,
    "Star Wars: Episode IV  A New Hope":True,
    "Star Wars: Episode V The Empire Strikes Back":True,
    "Star Wars: Episode VI Return of the Jedi":True,
    np.NaN: False    
}

for c in star_wars.columns[3:9]:
    star_wars[c] = star_wars[c].map(episodes)
```

```python
star_wars["Unnamed: 4"].value_counts(dropna=False) #check one column
```

We can also change the name of those columns to better reflect what they represent:

```python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
    })

star_wars.head()
```

### Cleaning the ranking columns

The next six columns ask the respondent to rank the Star Wars movies in order of least favorite to most favorite (1 = most favorite, 6 = least favorite). Each of the following columns can contain the value 1, 2, 3, 4, 5, 6 or NaN:
- `Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.` - how much the respondent liked Star Wars: Episode I The Phantom Menace
- `Unnamed: 10` - how much the respondent liked Star Wars: Episode II Attack of the Clones
- `Unnamed: 11` - how much the respondent liked Star Wars: Episode III Revenge of the Sith
- `Unnamed: 12` - how much the respondent liked Star Wars: Episode IV A New Hope
- `Unnamed: 13` - how much the respondent liked Star Wars: Episode V The Empire Strikes Back
- `Unnamed: 14` - how much the respondent liked Star Wars: Episode VI Return of the Jedi

To convert the columns to a numeric type and rename them:

```python
# to covert columns from 9 to 14, not including 15
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
    "Unnamed: 10":"ranking_2",
    "Unnamed: 11":"ranking_3",
    "Unnamed: 12":"ranking_4",
    "Unnamed: 13":"ranking_5",
    "Unnamed: 14":"ranking_6"
    })

star_wars.head()
```

### Finding the highest-ranked movie

To find the highest-ranked movie we can take the mean of each of the ranking columns:

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
import matplotlib.pyplot as plt
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

The more favorite the movie is the lower rank it has. We can see in the bar chart that the Original Trilogy (Episodes IV, V and VI) is ranked lower, thus being favoured more, than the Prequel Trilogy (Episodes I, II and III).

Also, the lowest ranked movie (= the most favored) is Star Wars: Episode V The Empire Strikes Back, and it's followed by Star Wars: Episode VI Return of the Jedi and Star Wars: Episode IV A New Hope.

The highest ranked movie of the Original Trilogy (Star Wars: Episode IV A New Hope) is still ranked lower than the highest ranked movie from the Prequel Trilogy (Star Wars: Episode I The Phantom Menace), meaning that the least favored movie from the Original Trilogy is still favored more than the most favored movie from the Prequel Trilogy.

Regarding the Prequel Trilogy, the movies are ranked: Star Wars: Episode I The Phantom Menace, Star Wars: Episode II Attack of the Clones and then Star Wars: Episode III Revenge of the Sith.

### Finding the most viewed movie

To find out how many respondents have seen each movie:

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

More respondents saw the Original Trilogy compared to the Prequel Trilogy. Only exception is the first movie from the Prequel Trilogy, which was seen by more respondents than the first movie from the Original Trilogy. 

Star Wars: Episode V The Empire Strikes Back was seen by most respondents and Star Wars: Episode III Revenge of the Sith was seen by least respondents.

The results are consistent with the rankings of the movies.

### Exploring the data by binary segments

We can split the dataframe into two groups based on a binary column. For example, we can split on the `Gender` column:

```python
males = star_wars[star_wars["Gender"]=="Male"]
females = star_wars[star_wars["Gender"]=="Female"]
```

To find the most viewed movies separately for each group:

```python
plt.bar(range(6), males[males.columns[3:9]].mean())
plt.title("Male")
plt.show()

plt.bar(range(6), females[females.columns[3:9]].mean())
plt.title("Female")
plt.show()
```

To find the highest-ranked movies separately for each group:

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.title("Male")
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean())
plt.title("Female")
plt.show()
```

Most viewed movies for males and females are Episode IV, Episode V and Episode I. Also more males have seen the Prequel Trilogy then the females.

The highest-ranked movie for males and females both is Episode IV. Also they both prefer the Original Trilogy over Prequel Trilogy. Also, first two movies of the Prequel Trilogy are ranked higher by females than by males.
